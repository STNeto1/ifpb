```python
# ==== QUESTÃO 4 ====
cars93 = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
def changeType(colname):
    if colname == 'Type':
        return 'CarType'
    return colname

def replaceChar(colname):
    return colname.replace('.', '_')

newColnames = list(map(changeType, cars93.columns))
newColnames = list(map(replaceChar, newColnames))

cars93.columns = newColnames
```

```python
# ==== QUESTÃO 5 ====

cars93.isnull()
```

```python
# ==== QUESTÃO 6 ====

cars93.isnull().sum().sort_values(ascending=False).head(1)
```

```python
# ==== QUESTÃO 8 ====

# A)

cols = list(df.columns)
cols[0] = 'c'
cols[2] = 'a'

df = df[cols]
```

```python
# B)

def change_col_names(df, name1, name2):
    cols = list(df)
    n1_idx = cols.index(name1)
    n2_idx = cols.index(name2)
    cols[n1_idx] = name2
    cols[n2_idx] = name1
    return df.reindex(columns=cols)

change_col_names(df, 'a', 'c')
```
